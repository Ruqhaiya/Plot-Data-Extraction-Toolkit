```python
import csv
import os
import numpy as np
import matplotlib.pyplot as plt

file_name = '3A_ICc_ICx'
image_dir = 'plot_images'
data_dir = 'plot_data_files'
output_dir = 'replotted_figures_paper_2'
os.makedirs(output_dir, exist_ok=True)

x_points = []
y_points = []

file_path = os.path.join(data_dir, f'{file_name}.csv')
with open(file_path, newline='') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        x, y = map(float, row)
        x_points.append(x)
        y_points.append(y)

x_points_icx = x_points[:20]
y_points_icx = y_points[:20]
x_points_icc_shell = x_points[20:]
y_points_icc_shell = y_points[20:]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 3.5))


# plotting ICc shell data points
ax1.scatter(x_points_icc_shell, y_points_icc_shell, edgecolor='black', facecolor='none', s=20, label='ICc shell (n=30)', marker='s')


# plotting ICx data points
ax1.scatter(x_points_icx, y_points_icx, edgecolor='black', facecolor='black', s=20, label='ICx (n=20)', marker='D')

ax1.set_xlim(1000, 9000)
ax1.set_ylim(0, 275)
ax1.set_xticks([1000, 3000, 5000, 7000, 9000])
ax1.set_yticks([50, 100, 150, 200, 250])
ax1.set_xlabel('Frequency (Hz)')
ax1.set_ylabel('Half-Peak Width (sec)')
ax1.set_title('Re-plotted figure')
ax1.legend()

#displaying the original plot here
original_plot_path = os.path.join(image_dir, f'{file_name}.png')
original_img = plt.imread(original_plot_path)
ax2.imshow(original_img)
ax2.axis('off')
ax2.set_title(f'Original Plot ({file_name})')

plt.tight_layout()
plt.savefig(os.path.join(output_dir, f'{file_name}.png'))
plt.show()
plt.close()

print("Success!!!")
```

# Labeling

```python
import csv
import os

file_name = '3A_ICc_ICx'
data_dir = 'plot_data_files'
output_file = '3A_ICc_ICx_labeled' + '.csv'
output_dir = 'plot_data_files'
os.makedirs(output_dir, exist_ok=True)

x_points = []
y_points = []
labels = []

file_path = os.path.join(data_dir, f'{file_name}.csv')
with open(file_path, newline='') as csvfile:
    reader = csv.reader(csvfile)
    for index, row in enumerate(reader):
        x, y = map(float, row)
        x_points.append(x)
        y_points.append(y)
        if index < 20:
            labels.append('ICx')
        else:
            labels.append('ICc')

output_file_path = os.path.join(output_dir, output_file)
with open(output_file_path, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['X', 'Y', 'Label'])
    for x, y, label in zip(x_points, y_points, labels):
        writer.writerow([x, y, label])

print(f"Modified CSV file saved to {output_file_path}")
```
