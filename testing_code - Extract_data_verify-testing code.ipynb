```python
import csv
import pandas as pd
import matplotlib.pyplot as plt

def round_val(value, threshold=0.08):
    nearest_half = round(value * 2) / 2
    if abs(value - nearest_half) <= threshold:
        return nearest_half
    return value
```

# B 40

```python
x_points = []
y_points = []

# extracting data from CSV
file_path = 'B_plot_pg40_datapoints.csv'
with open(file_path, newline='') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        x, y = map(float, row)
        x_points.append(x)
        y_points.append(y)

# rounding x-axis values
x_points_rounded = [round_val(x) for x in x_points]
# print(x_points_rounded)


#saving rounded data to a new CSV file
rounded_data_1 = list(zip(x_points_rounded, y_points))
output_file_path_1 = 'B_plot_pg40_datapoints_rounded.csv'
with open(output_file_path_1, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(rounded_data_1)
    
x_scale = (2.5, 6.5)
y_scale = (2.5, 22.5)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# plotting original data
ax1.scatter(x_points, y_points, edgecolor='black', facecolor='none', s=15)
ax1.set_xlim(x_scale)
ax1.set_ylim(y_scale)
ax1.set_xticks([3, 4, 5, 6])
ax1.set_yticks(range(5, 25, 5))
ax1.set_title('Original Values')
ax1.set_xlabel('Mediolateral Position (mm)')
ax1.set_ylabel('Response Latency (ms)')

# plottng rounded data
ax2.scatter(x_points_rounded, y_points, edgecolor='black', facecolor='none', s=15)
ax2.set_xlim(x_scale)
ax2.set_ylim(y_scale)
ax2.set_xticks([3, 4, 5, 6])
ax2.set_yticks(range(5, 25, 5))
ax2.set_title('Rounded Values')
ax2.set_xlabel('Mediolateral Position (mm)')
ax2.set_ylabel('Response Latency (ms)')

plt.tight_layout()
plt.show()
```
