```python
import csv
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy import stats

def fit_regression_line(x, y):
    x = np.array(x).reshape(-1, 1)
    y = np.array(y)
    model = LinearRegression().fit(x, y)
    y_pred = model.predict(x)
    slope, intercept, r_value, p_value, std_err = stats.linregress(x.flatten(), y)
    return model, y_pred, slope, intercept, r_value, p_value, std_err, len(x)

file_names = [
    "40_B",
    "42_A",
    "42_B",
    "45_A",
    "45_B",
    "45_C",
    "47_A",
    "47_B",
    "47_C"
]

#folder that contains the rounded plot data files
rounded_data_dir = 'rounded_data_files'

#folder of original plot images
image_dir = 'plot_images'

#folder to save the re-plotted figures
output_dir = 'replotted_figures_rounded_4_5mm_with_regression_lines'
os.makedirs(output_dir, exist_ok=True)

# Dicts for scales, ticks, and axis labels
scales = {
    "40_B": ((2.5, 6.5), (2.5, 22.5)),
    "42_A": ((2.5, 6.5), (0.5, 10.25)),
    "42_B": ((2.5, 6.5), (0, 8)),
    "45_A": ((2.5, 6.5), (-300, 150)),
    "45_B": ((2.5, 6.5), (25, 300)),
    "45_C": ((2.5, 6.5), (-32, 90)),
    "47_A": ((2.5, 6.5), (-25, 25)),
    "47_B": ((2.5, 6.5), (8.5, 65)),
    "47_C": ((2.5, 6.5), (0, 0.7))
}

yticks = {
    "40_B": [5, 10, 15, 20],
    "42_A": [2, 4, 6, 8, 10],
    "42_B": [2, 4, 6],
    "45_A": [-200, -100, 0, 100],
    "45_B": [50, 100, 150, 200, 250],
    "45_C":  [-20, 0, 20, 40, 60, 80],
    "47_A": [-20, -10, 0, 10, 20],
    "47_B": [10, 20, 30, 40, 50, 60],
    "47_C": [0.2, 0.4, 0.6]
}

axis_labels = {
    "40_B": ('Mediolateral Position (mm)', 'Response Latency (ms)'),
    "42_A": ('Mediolateral Position (mm)', 'Best Frequency (kHz)'),
    "42_B": ('Mediolateral Position (mm)', 'Frequency Tuning Width (kHz)'),
    "45_A": ('Mediolateral Position (mm)', 'Best ITD (us)'),
    "45_B": ('Mediolateral Position (mm)', 'ITD tuning Width (us)'),
    "45_C": ('Mediolateral Position (mm)', 'Sidepeak Supression (%)'),
    "47_A": ('Mediolateral Position (mm)', 'Best ILD (dB)'),
    "47_B": ('Mediolateral Position (mm)', 'ILD tuning Width (dB)'),
    "47_C": ('Mediolateral Position (mm)', 'ILD Symmetry Index')
}

# Looping through each plot file to create re-plotted figures with regression lines
for plot_file in file_names:
    x_points = []
    y_points = []

    file_path = os.path.join(rounded_data_dir, f'{plot_file}_rounded.csv')
    with open(file_path, newline='') as csvfile:
        reader = csv.reader(csvfile)
        for row in reader:
            x, y = map(float, row)
            x_points.append(x)
            y_points.append(y)

    x_scale, y_scale = scales[plot_file]
    xticks_vals = [3, 4, 5, 6]
    yticks_vals = yticks[plot_file]
    x_label, y_label = axis_labels[plot_file]

    # regression line for all data
    model_all, y_pred_all, slope_all, intercept_all, r_value_all, p_value_all, std_err_all, n_all = fit_regression_line(x_points, y_points)

    #filtering data for mediolateral position between 4 and 5 mm
    filtered_indices = [i for i, x in enumerate(x_points) if 4 <= x <= 5]
    x_points_filtered = [x_points[i] for i in filtered_indices]
    y_points_filtered = [y_points[i] for i in filtered_indices]

    # regression line for filtered data
    model_filtered, y_pred_filtered, slope_filtered, intercept_filtered, r_value_filtered, p_value_filtered, std_err_filtered, n_filtered = fit_regression_line(x_points_filtered, y_points_filtered)

    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 4.5))

    # plotting original data with regression line
    ax1.scatter(x_points, y_points, edgecolor='black', facecolor='none', s=15)
    ax1.plot(x_points, y_pred_all, color='red', linestyle='--')
    ax1.set_xlim(x_scale)
    ax1.set_ylim(y_scale)
    ax1.set_xticks(xticks_vals)
    ax1.set_yticks(yticks_vals)
    ax1.set_title(f'Rounded Data with Regression Line ({plot_file})')
    ax1.set_xlabel(x_label)
    ax1.set_ylabel(y_label)

    stats_text_all = (f'r: {r_value_all:.3f}\np: {p_value_all:.4f}\nn: {n_all}')
    ax1.text(0.05, 0.95, stats_text_all, transform=ax1.transAxes, fontsize=12,
             verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.5))

    # plotting filtered data with regression line
    ax2.scatter(x_points_filtered, y_points_filtered, edgecolor='black', facecolor='none', s=15)
    ax2.plot(x_points_filtered, y_pred_filtered, color='blue', linestyle='--')
    ax2.set_xlim(x_scale)
    ax2.set_ylim(y_scale)
    ax2.set_xticks(xticks_vals)
    ax2.set_yticks(yticks_vals)
    ax2.set_title(f'Filtered Data (4-5mm) with Regression Line ({plot_file})')
    ax2.set_xlabel(x_label)
    ax2.set_ylabel(y_label)

    stats_text_filtered = (f'r: {r_value_filtered:.3f}\np: {p_value_filtered:.4f}\nn: {n_filtered}')
    ax2.text(0.05, 0.95, stats_text_filtered, transform=ax2.transAxes, fontsize=12,
             verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.5))

    # Displaying the original plot
    original_plot_path = os.path.join(image_dir, f'{plot_file}.png')
    original_img = plt.imread(original_plot_path)
    ax3.imshow(original_img)
    ax3.axis('off')
    ax3.set_title(f'Original Plot ({plot_file})')

    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, f'{plot_file}_regression.png'))
    plt.show()
    plt.close()

print("Success!!!")
```
