```python
import csv
import os
import numpy as np
import matplotlib.pyplot as plt

file_name = '10_A_dist_peaks_ICc' 
image_dir = 'plot_images'
data_dir = 'plot_data_files'
output_dir = 'replotted_figures_paper_3'
os.makedirs(output_dir, exist_ok=True)

x_points = []
y_points = []

file_path = os.path.join(data_dir, '10_A_dist_peaks_ICc.csv')
with open(file_path, newline='') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        x, y = map(float, row)
        x_points.append(x)
        y_points.append(y)

print(len(x_points), len(y_points))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 3.5))

ax1.scatter(x_points, y_points, edgecolor='black', facecolor='black', s=40, marker='^')

ax1.set_xlim(0, 10)
ax1.set_ylim(-0.75, 0.75)
ax1.set_xticks([0, 2, 4, 6, 8, 10])
ax1.set_yticks([-0.75, -0.5, -0.25, 0, 0.25, 0.5, 0.75])
ax1.set_xlabel('Frequency (kHz)')
ax1.set_ylabel('Occurrences')
ax1.set_title('Re-plotted figure')

original_plot_path = os.path.join(image_dir, f'{file_name}.png')
original_img = plt.imread(original_plot_path)
ax2.imshow(original_img)
ax2.axis('off')
ax2.set_title(f'Original Plot ({file_name})', fontsize=14)

plt.tight_layout()
plt.savefig(os.path.join(output_dir, f'{file_name}.png'))
plt.show()
plt.close()

print("Success!!!")
```

# Adding grey area

```python
import csv
import os
import numpy as np
import matplotlib.pyplot as plt

file_name = '10_A_dist_peaks_ICc' 
image_dir = 'plot_images'
data_dir = 'plot_data_files'
output_dir = 'replotted_figures_paper_3'
os.makedirs(output_dir, exist_ok=True)

x_points = []
y_points = []

file_path = os.path.join(data_dir, '10_A_dist_peaks_ICc.csv')
with open(file_path, newline='') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        x, y = map(float, row)
        x_points.append(x)
        y_points.append(y)

print(len(x_points), len(y_points))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 3.5))

ax1.scatter(x_points, y_points, edgecolor='black', facecolor='black', s=40, marker='^')

#adding grey area
ax1.axhline(y=0.25, color='grey', linewidth=2)
ax1.axhline(y=-0.25, color='grey', linewidth=2)
ax1.fill_between([0, 10], -0.25, 0.25, color='grey', alpha=0.2)

ax1.set_xlim(0, 10)
ax1.set_ylim(-0.75, 0.75)
ax1.set_xticks([0, 2, 4, 6, 8, 10])
ax1.set_yticks([-0.75, -0.5, -0.25, 0, 0.25, 0.5, 0.75])
ax1.set_xlabel('Frequency (kHz)')
ax1.set_ylabel('Occurrences')
ax1.set_title('Re-plotted figure')

original_plot_path = os.path.join(image_dir, f'{file_name}.png')
original_img = plt.imread(original_plot_path)
ax2.imshow(original_img)
ax2.axis('off')
ax2.set_title(f'Original Plot ({file_name})', fontsize=14)

plt.tight_layout()
plt.savefig(os.path.join(output_dir, f'{file_name}.png'))
plt.show()
plt.close()

print("Success!!!")
```
